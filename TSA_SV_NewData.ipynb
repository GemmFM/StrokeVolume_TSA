```python
# Funci√≥n desviaci√≥n standard

cv = lambda x:np.std(x, ddof = 1)
```

```python
# Coeficientes de Variaci√≥n

# Controles
cv_SV_C_diff = [cv(SV_C1_diff), cv(SV_C2_diff), cv(SV_C3_diff), cv(SV_C4_diff),
           cv(SV_C5_diff), cv(SV_C6_diff), cv(SV_C7_diff), cv(SV_C8_diff),
           cv(SV_C9_diff), cv(SV_C10_diff), cv(SV_C11_diff), cv(SV_C12_diff),
           cv(SV_C13_diff), cv(SV_C14_diff)]

# Angiotensinas
cv_SV_A_diff = [cv(SV_A1_diff), cv(SV_A2_diff), cv(SV_A3_diff), cv(SV_A4_diff),
           cv(SV_A5_diff), cv(SV_A6_diff), cv(SV_A7_diff), cv(SV_A8_diff),
           cv(SV_A9_diff), cv(SV_A10_diff)]
print(cv_SV_C_diff)
print(cv_SV_A_diff)
```

```python
# Gr√°fica promedio y desviaci√≥n

m1 = [np.mean(cv_SV_C_diff),np.mean(cv_SV_A_diff)]
d1 = [np.std(cv_SV_C_diff),np.std(cv_SV_A_diff)]
e1 = [d1[0]/np.sqrt(14), d1[1]/np.sqrt(9)]

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.style.use('seaborn-whitegrid')

index = np.arange(2)
width = 0.7
labels = ('Controls', 'Ang_II')

fig, ax = plt.subplots()

ax.p1 = plt.bar(index, m1, width=width, color="blue",
               linewidth=2, zorder=5, alpha=0.5)

ax.errs = plt.errorbar(index, m1, yerr=e1, fmt='.k')

plt.xticks(index, labels)

plt.ylabel('Average of Standard Desviation 1a diff')
plt.show()
```
